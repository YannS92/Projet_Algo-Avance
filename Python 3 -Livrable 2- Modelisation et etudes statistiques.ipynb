### 3.2.1 Gnration de la matrice d'adjacence <a class="anchor" id="matriceAdjacence"></a>

Nous avons implemanter une fonction nomme "generer_matrice_adjacence()" qui permet comme son nom l'indique gnrer notre matrice d'adjacence, elle prend en paramtre deux arguments le nombre de villes et le type du graphe (graphe complet / graphe non-complet).

```python
import numpy as np
import random

random.seed(10)

nb_villes = 10                                           # On definit le nombre de ville

def generer_matrice_adjacence(nb_villes,typeDeGraphe):   # Cette fonction prend en paramtre deux arguments le nombre de ville et le type dgu graphe   
    matrice = np.empty((nb_villes, nb_villes))        # On initialise la matrice           
    
    if(typeDeGraphe == 'Complete'):          # 1er Cas : Graphe complet     
        
        for i in range(nb_villes):        # On parcourt notre matrice symtrique   
            for j in range(i,nb_villes):    
                valeur = 1 if i!=j else 0    # Dans le cas ou les index i et j sont gaux (position diagonal) on met la valeur  0
                matrice[i][j] = valeur       # Toutes les valeurs (hors diagonal) sont gales  1
                matrice[j][i] = valeur
    else:
         for i in range(nb_villes):       # 2eme ca : Graphe non complet 
            for j in range(i,nb_villes):
                valeur = random.randint(0,1)  if i!=j else 0   
                matrice[i][j] = valeur
                matrice[j][i] = valeur
    return matrice

matriceAdjacence = generer_matrice_adjacence(nb_villes, "Not-Complete")
print("Matrice d'adjacence graphe non-complet : \n",matriceAdjacence)

print("\n \n")     

matriceAdjacence = generer_matrice_adjacence(nb_villes, "Complete")
print("Matrice d'adjacence graphe complet : \n",matriceAdjacence)
```

### 3.2.3 Gnration de la liste des voisins de chaque sommet en se basant sur la matrice d'adjacence <a class="anchor" id="lister_voisins"></a>

Nous avons dvelopp la fonction "lister_voisins()" qui prend en paramtre notre matrice d'adjacence et qui permet d'assoccier  chaque sommet sa liste de voisins.

```python
def lister_voisins(matriceAdjacence):       # Cette fonction prend en paramtre une matrice d'adjacence
    taille = len(matriceAdjacence.T) 
    voisins = []
    i=0
    while (i < taille):                     # On parcourt la matrice d'adjacence
        for j in range(taille):
            if(matriceAdjacence[i][j] != 0 and matriceAdjacence[j][i] != 0):   # si la valeur  la position [i][j] est differente de zro alors c'est un voisin du sommet
                voisins.append(j)       

        print("Les voisins du sommet", i, "sont :",voisins)  # Afficher les voisins de chaque sommet
        voisins.clear()
        i+=1
        
lister_voisins(matriceAdjacence)
```

### 3.2.4 Gnration de la matrice pondre<a class="anchor" id="matricePondere"></a>

Nous avons dvelopp la fonction "generer_matrice_pondere()" qui comme son nom l'indique gnre une matrice pondre, cette fonction prend en paramtre une matrice d'adjacence.
Le but est de remplacer les valeurs gales  1 par des valeurs entires positives alatoires qui reprsenteront la distance.

```python
random.seed(10)

marge_distance_max = 200

def generer_matrice_pondere(matrice):
    for i in range(len(matrice.T)):             # On parcourt la matrice d'ajdacence 
        for j in range(i,len(matrice.T)):
            if(matrice[i][j] != 0):
                valeur = random.randint(0, marge_distance_max)     # On crase les valeurs gale  1 par des valeurs entire positives alatoire qui reprsentent la distance 
                matrice[i][j] = valeur
                matrice[j][i] = valeur
    return matrice

graphe  = generer_matrice_pondere(matriceAdjacence)
print("Matrice pondre : \n\n", graphe)
```

### 3.2.5 Gnration de la solution initiale <a class="anchor" id="SolInit"></a>

Nous avons dvelopp la fonction "generate_solution()" qui prend en paramtre une matrice pondre, un nombre de vhicules Min et Max, cette fonction va gnrer une structure dictionnaire qui aura comme cls les camions et comme valeur des listes de chemins ce dictionnaire reprsentera notre solution initiale.


```python
minVehicule = 2
maxVehicule = 5

def generate_solution(matrice,minVehicule,maxVehicule):
    sol = {}
    matriceSearch = [i for i in range(len(matrice))]
    matriceSearch.remove(0)
    i = 0
    nbvehicule = random.randint(minVehicule,maxVehicule)
    while (len(matriceSearch) > 0):
        way = []
        way.append(0)
        for j in range(0,(len(matrice)//nbvehicule)) :
            if len(matriceSearch) > 0 :
                value = random.choice(matriceSearch)
                way.append(value)
                matriceSearch.remove(value)
            else: 
                break
        way.append(0)
        sol[i] = way
        i += 1
    return sol,nbvehicule

sol,nbvehicule=generate_solution(graphe,minVehicule,maxVehicule)  
print("La solution Initiale So : \n" ,sol)
```

### 3.2.6 Calcul du cout d'un chemin<a class="anchor" id="cout"></a>

```python
# Fonction qui prend en paramtre une solution et une matrice pondre et qui renvoie le cout total de la solution 
def weightSol(sol,graphe):
    return sum(graphe[sol[i][j]][sol[i][j+1]] for i in sol for j in range(len(sol[i])-1))

cout = weightSol(sol,graphe)
print("Le cout de la solution : ", sol ,"est : ", cout)
```

### 3.2.7 Gnration du voisinage d'une solution<a class="anchor" id="voisinage"></a>

Aprs avoir gnrer la solution initiale (So), l'tape suivante consiste  crer la liste des solutions candidates  partir de la solution actuelle S (solution initiale  l'itration 0), nous appelons ces solutions le voisinage de (So). 
Pour trouver les solutions voisines  partir de la solution actuelle So, nous avons dvelopp la fonction de voisinage "voisinage()" qui prend en paramtre la solution initiale So et qui renvoie pour chaque camion K une liste de chemin possible en effectuant un mouvement d'change (swap).

```python
import pprint

# Gnre les voisins valides d'une solution
def reformSolution(solutionConcatenate:list):
    idx_list = [idx + 1 for idx in range(len(solutionConcatenate)-1) if solutionConcatenate[idx] == 0 and solutionConcatenate[idx+1]== 0 ]
    if len(idx_list) > 0:
        solutionTransform = [solutionConcatenate[i: j] for i, j in zip([0] + idx_list, idx_list +  ([len(solutionConcatenate)] if idx_list[-1] != len(solutionConcatenate) else []))]
        solutionFinal = [val for val in solutionTransform if val != [0] ]
        solution = {truck:way for truck,way in enumerate(solutionFinal)}
    else:
        solution = {0:solutionConcatenate}
    return solution

def voisinage(sol:dict):
    voisinage = []
    solution = []
    for truckWay in sol.values():
        solution += truckWay
    for i in range (1,len(solution)-1):
        for n in range(1,len(solution)-i-2):
            neighborPath=solution.copy()
            if solution[i] == 0:
               continue
            elif solution[i+n] == 0:
                index = n + 2   
                neighborPath = solution[:i]+[solution[i],solution[i+index]]+solution[i+1:]
                neighborPath.pop(i+index+1)
                voisinage.append(reformSolution(neighborPath))
                neighborPath = solution[:i+index]+[solution[i+index],solution[i]]+solution[i+index+1:]
                neighborPath.pop(i)
                voisinage.append(reformSolution(neighborPath))
            else:
                neighborPath[i], neighborPath[i+n] = neighborPath[i+n], neighborPath[i]
                voisinage.append(reformSolution(neighborPath))
    return voisinage

Voisinage = voisinage(sol)
pp = pprint.PrettyPrinter(depth=3,sort_dicts=False)
pp.pprint(Voisinage)
```

#### 3.3.Implmentation de la mtaheuristique "Recherche Tabou"<a class="anchor" id="tabou"></a>

En partant d'une solution quelconque x appartenant  l'ensemble de solutions X , nous nous dirigeons vers une solution s(x) appartenant au voisinage S(x) de x. 
L'algorithme explore donc itrativement l'espace de solutions X . Afin de choisir le meilleur voisin s(x) , l'algorithme value la fonction objectif f en chaque point s(x), et retient le voisin qui amliore la valeur de f , ou celui qui la dgrade le moins.

```python
def tabuSearch(sol, tabuLength, iterMax, graphe):

    nbIter = 0
    tabuList = list()

    initialElement = sol
    # variables solutions pour la recherche du voisin optimal non tabou
    currentElement = initialElement
    bestNeighbor=currentElement
    bestNeighborGlobal=currentElement

    # variables valeurs pour la recherche du voisin optimal non tabou
    bestCost=100000
    bestCostGlobal=100000

    # variables pour l'affichage
    nbTabou=0
    bestGlobalFound=0
    
    # liste des solutions courantes et des meilleures trouves, pour afficher la trajectoire
    paths=list()
    bestPaths=list()
    
    while (nbIter<iterMax):
        nbIter += 1       
        bestCost=100000

        # on parcours tous les voisins de la solution courante
        for neighbor in voisinage(currentElement):
            if weightSol(neighbor,graphe) < bestCost:
                if neighbor not in tabuList:
                    bestCost = weightSol(neighbor,graphe)
                    bestNeighbor = neighbor
          
        # on met a jour la meilleure solution rencontree depuis le debut
        if bestCost<bestCostGlobal:
            bestGlobalFound+=1
            bestNeighborGlobal=bestNeighbor
            bestCostGlobal=bestCost
            #print("Meilleur global trouv ! : " + str(bestNeighborGlobal) + " avec une valeur de :" + str(bestCostGlobal))
      
        bestPaths.append(bestCostGlobal)
        
        # on passe au meilleur voisin non tabou trouve     
        currentElement=bestNeighbor.copy()
        paths.append(bestCost)
        
        # on met a jour la liste tabou
        tabuList.append(bestNeighbor)

        # on supprime la solution la plus ancienne si la liste tabou  atteint sa taille maximale
        if len(tabuList) > tabuLength:
            del tabuList[0]

    # On insre  l'lment initiale la ville d'origine
    return bestNeighborGlobal, paths, bestPaths, initialElement
```

```python
import time
import matplotlib.pyplot as plt

# Gnration de la matrice pondre
graphe = generer_matrice_pondere(matriceAdjacence)

# Initialisation de nombre d'itration Max 
iterMaximum=100
# Modifier ce paramtre pour modifier la taille maximum de la liste tabou
tabuMax=len(graphe)
# Modifier ce paramtre pour activer ou non l'utilisation du plus proche voisin comme chemin initial
useNearestNeighbor=True
# Modifier ce paramtre pour changer la ville de dpart
departureCity=0

startTime = time.time()
sol, paths, bestPaths, initialPath=tabuSearch(sol, tabuLength=tabuMax, iterMax=iterMaximum, graphe=graphe)
execTime = (time.time() - startTime)

print("La valeur du chemin initial est :",str(weightSol(initialPath,graphe)))
print("\n---------------------------------------------")
#print("\nLe meilleur chemin trouv est : " + str(resNearestNeighbor(sol)))
print("\nSa valeur est de : " + str(weightSol(sol,graphe)) + " mn, soit " + str("{:.2f}".format(weightSol(sol,graphe)/60)) + " h")
print("\nTemps d'execution : " + str(execTime) + " secondes\n\n")

# Affichage des rsultats sur le graphique
plt.xlabel("Nombre d'itrations", fontsize=16)
plt.ylabel("Valeur de solution", fontsize=16)

res = plt.plot(range(iterMaximum), paths,label='Meilleure valeur pour chaque itration')
res = plt.plot(range(iterMaximum), bestPaths,label='Meilleure valeur globale')

plt.legend()
font2 = {'size':25}
plt.title("\nExploration de la zone de recherche par notre algorithme\n", loc = 'center',fontdict = font2)
```

On observe des oscillations entre des solutions proches, dont l'une est meilleure que l'autre. Cette solution est un optimum local, et l'algorithme semble tre prisonnier de la "valle" de cet optimum local, et qui s'appelle son <em>bassin d'attraction</em>. C'est ce phnomne qui justifie d'effectuer des tapes de diversification, afin de s'chapper de ce bassin d'attraction. La conclusion est qu'il faut renforcer la diversification. La premire ide qui vient  l'esprit pour remdier  ce problme consiste  agrandir la taille de la liste tabou.

Si on agrandit la taille de la liste tabou sans augmenter le nombre d'itrations, la liste tabou n'aura pas le temps de se remplir. Il faut donc aussi augmenter le nombre d'itrations

Si notre solution est loin de la solution, alors la solution n'est pas correcte. L'algorithme suivant cre une borne infrieur qui est donn via le traitement de la fonction objective tout en prenant les donnes des variables et des contraintes.AC

```python
import numpy as np
from pulp import *

def borne_inferieur():
    # variables
    x = [[[pulp.LpVariable("x%s_%s,%s"%(i,j,k), cat="Binary") if i != j else None for k in range(nbvehicule-1)] for j in range(nb_villes)] for i in range(nb_villes)]
    # probleme
    prob = LpProblem("vrp", LpMinimize)

    # fonction objective
    cost = lpSum(graphe[i][j] * x[i][j][k] if i != j else 0 for k in range(nbvehicule-1) for j in range(nb_villes) for i in range (nb_villes))
    prob += cost

    # contrainte
    for j in range(nb_villes):
        prob +=lpSum(x[i][j][k] if i != j else 0 for i in range(0,nb_villes) for k in range(nbvehicule-1)) == 1
    
    for k in range(nbvehicule-1):
        prob+= lpSum(x[0][j][k] for j in range(nb_villes)) == 1
        prob+= lpSum(x[i][0][k] for i in range(nb_villes)) == 1

    for k in range(nbvehicule-1):
        for j in range(nb_villes):
            prob+= lpSum(x[i][j][k] if i != j else 0 
                                  for i in range(nb_villes)) -  lpSum(x[j][i][k] for i in range(nb_villes)) == 0
    prob.solve()
    return value(prob.objective)
```
